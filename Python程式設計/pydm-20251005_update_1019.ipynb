<b>Tech</b>

    <li>Loop</li>

<b>Tasks</b>
<ol>
    <li>隨機各發一張撲克牌給 player1 與 player2，撲克牌結構範例為 ('SPADE', '3') 代表 黑桃3</li>
    <li>比較 player1 與 player2 勝負
        <ol>
            <li type="a">先比數字，數字大小依序為 '2' > 'A' > 'K' > 'Q' > 'J' > '10' > '9' > '8' > '7' > '6' > '5' > '4' > '3'，較大者獲勝
            <li type="a">若雙方數字相同則比較花色，花色大小依序為 'SPADE' > 'HEART' > 'DIAMOND' > 'CLUB'，較大者獲勝
        </ol>
    </li>
    <li>輸出 player1 與 player2 勝負</li>
</ol>

```python
import random

SUITS = ['SPADE', 'HEART', 'DIAMOND', 'CLUB']
NUMBERS = ['2', 'A', 'K', 'Q', 'J', '10', '9', '8', '7', '6', '5', '4', '3']

# init 52 cards
cards = [(s, n) for s in SUITS for n in NUMBERS]

# shuffle
random.shuffle(cards)

# deal
p1, p2 = cards[:2]
cards = cards[2:]
print(p1, p2)

# compare
suits_w = {s: w for w, s in enumerate(SUITS)}
numbers_w = {n: w for w, n in enumerate(NUMBERS)}

p1_n_w, p1_s_w = numbers_w[p1[1]], suits_w[p1[0]]
p2_n_w, p2_s_w = numbers_w[p2[1]], suits_w[p2[0]]

if p1_n_w < p2_n_w:  # comprae numbers
    print(f'p1 {p1} win')
elif p1_n_w > p2_n_w:
    print(f'p2 {p2} win')
elif p1_s_w < p2_s_w:  # compare suits
    print(f'p1 {p1} win')
elif p1_s_w > p2_s_w:
    print(f'p2 {p2} win')
else:
    print('tie')
```

```python
import random

SUITS = ['SPADE', 'HEART', 'DIAMOND', 'CLUB']
NUMBERS = ['2', 'A', 'K', 'Q', 'J', '10', '9', '8', '7', '6', '5', '4', '3']

# init 52 cards
cards = [(s, n) for s in SUITS for n in NUMBERS]

# shuffle
random.shuffle(cards)

# deal
p1, p2 = cards[:2]
cards = cards[2:]
print(p1, p2)

# compare by list order
p1_n_w, p1_s_w = NUMBERS.index(p1[1]), SUITS.index(p1[0])
p2_n_w, p2_s_w = NUMBERS.index(p2[1]), SUITS.index(p2[0])

if p1_n_w < p2_n_w:  # comprae numbers
    print(f'p1 {p1} win')
elif p1_n_w > p2_n_w:
    print(f'p2 {p2} win')
elif p1_s_w < p2_s_w:  # compare suits
    print(f'p1 {p1} win')
elif p1_s_w > p2_s_w:
    print(f'p2 {p2} win')
else:
    print('tie')
```

```python
nums = [40, 30, 60, 50, 20]

for j in range(len(nums)-1):
    for i in range(len(nums)-1-j):
        if nums[i] > nums[i+1]:
            nums[i], nums[i+1] = nums[i+1], nums[i] #swap
nums
```

```python
nums = [40, 30, 60, 50, 20]

def bubble(nums, reverse=False):
    if reverse:
        for j in range(len(nums)-1):
            for i in range(len(nums)-1-j):
                if nums[i] < nums[i+1]:
                    nums[i], nums[i+1] = nums[i+1], nums[i] #swap
    else:
        for j in range(len(nums)-1):
            for i in range(len(nums)-1-j):
                if nums[i] > nums[i+1]:
                    nums[i], nums[i+1] = nums[i+1], nums[i] #swap

bubble(nums, True)
print(nums)
```
